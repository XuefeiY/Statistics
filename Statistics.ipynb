## 2. Statistical Tests
* Parametric Tests & Non-Parametric Test
  * Z Test 
  * T Test 
  * ANOVA
  * Chi Square Test
  * F Test
  
Z-test implies a hypothesis test which ascertains if the means of two datasets are different from each other when variance is given.

T-test refers to a univariate hypothesis test based on t-statistic, wherein the mean is known, and population variance is approximated from the sample.
